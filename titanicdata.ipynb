Data Preprocessing


```python
titanic_df.isnull().sum()
```

```python
titanic_df['age'].fillna(value=titanic_df['age'].mean(),inplace=True)
```

```python
titanic_df.isnull().sum()
```

```python
titanic_df.head(20)
```

```python
titanic_df['deck'].fillna(value=titanic_df['deck'].mode()[0],inplace=True)
```

```python
titanic_df.isnull().sum()
```

```python
titanic_df.head(10)
```

```python
titanic_df.dropna(inplace=True)
```

```python
titanic_df.isnull().sum()
```

```python
titanic_df.head(20)
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
titanic_df.info()
```

```python
encoder=LabelEncoder()
```

```python
titanic_df['sex']=encoder.fit_transform(titanic_df['sex'])
```

```python
titanic_df
```

```python
titanic_df['embarked']=encoder.fit_transform(titanic_df['embarked'])
```

```python
titanic_df['who']=encoder.fit_transform(titanic_df['who'])
```

```python
titanic_df['embark_town']=encoder.fit_transform(titanic_df['embark_town'])
```

```python
titanic_df['alive']=encoder.fit_transform(titanic_df['alive'])
```

```python
titanic_df['class']=encoder.fit_transform(titanic_df['class'])
titanic_df['deck']=encoder.fit_transform(titanic_df['deck'])
```

```python
titanic_df.drop(['class','who','adult_male','embark_town','alive'],axis=1,inplace=True)
```

```python
titanic_df
```

```python
x=titanic_df.drop(['survived'],axis=1)
y=titanic_df.survived
```

```python
print(x)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=123)
```

```python
x_train.shape,x_test.shape,y_train.shape,y_test.shape
```

```python
decision_model=DecisionTreeClassifier(max_depth=4)
```

```python
decision_model.fit(x_train,y_train)
```

```python
y_pred=decision_model.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score, precision_score, f1_score, recall_score
```

```python
accuracy_score(y_test,y_pred)
```

```python
precision_score(y_test,y_pred)
```

```python
f1_score(y_test,y_pred)
```

```python
recall_score(y_test, y_pred)
```

```python
from sklearn.metrics import roc_auc_score
```

```python
roc_auc_score(y_test, y_pred)
```
